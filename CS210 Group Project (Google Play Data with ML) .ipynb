# k-Nearest Neighbors

```python
# Our features and label data are ready, we can proceed to introduce them
# to k-NN algorithm. But firstly, we should create train and test sets from the 
# original datasets.
from sklearn.model_selection import train_test_split

X = df[["Rating", "Size"]]
y = df["Installs"]

# 75% for training and 25% for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.25, random_state = 42)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Let's try with k = 5
knn = KNeighborsClassifier(n_neighbors=37)
knn.fit(X_train, y_train)
```

```python
from sklearn import metrics
y_pred = knn.predict(X_test)
print("Accuracy of the model with k = 37 :",metrics.accuracy_score(y_test, y_pred))
```

```python
accuracy_list = []
for k in range (1, 101):
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy_list.append(metrics.accuracy_score(y_test, y_pred))
plt.figure(figsize=(12, 6))  
plt.plot(range(1, 101), accuracy_list, color='black', linestyle='dashed', marker='o',  
         markerfacecolor='red', markersize=10)
plt.title('Accuracy of the Model w/ k-Value')  
plt.xlabel('k-Value')  
plt.ylabel('Accuracy') 
plt.show()
```

According to the graph, k=37 has the highest accuracy, so we chose k=37.

# Normalizing the Data:

```python
dfnorm = df[["Rating", "Size", "Installs"]]
normalized_df = (dfnorm-dfnorm.mean()) / dfnorm.std() #Normalizing the data
X_norm = normalized_df[["Rating", "Size"]]
y = df["Installs"]

# 75% for training and 25% for testing
X_train, X_test, y_train, y_test = train_test_split(X_norm, y, test_size= 0.25, random_state = 42)
```

```python
knn = KNeighborsClassifier(n_neighbors=41)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

print("Accuracy of the model with k = 37 and using normalized features: ",metrics.accuracy_score(y_test, y_pred))
```

```python
#import required packages
from sklearn import neighbors
from sklearn.metrics import mean_squared_error 
from math import sqrt
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.neighbors import KNeighborsRegressor
accuracy_list = []
for k in range (1, 101):
    knn = KNeighborsRegressor(n_neighbors = k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    error = sqrt(mean_squared_error(y_test,y_pred))
    accuracy_list.append(error)

plt.figure(figsize=(12, 6))  
plt.plot(range(1, 101), accuracy_list, color='black', linestyle='dashed', marker='o',  
         markerfacecolor='red', markersize=10)
plt.title('Error rate of the Model w/ k-Value')  
plt.xlabel('k-Value')  
plt.ylabel('Error') 
plt.show()

min_error = min(accuracy_list)
print("Minimum error rate: ",min_error)
```

The highest accuracy is at k= 5 with the error rate of 33385751.282507733, so we chose k=5.
